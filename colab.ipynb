## Name of the dataset
We will use the **High School Longitudinal Study of 2009 (HSLS:09)**.
HSLS:09 is a longitudinal study implemented and managed by the U.S. National Center for Education Statistics (NCES) that collected responses from the 9th graders in 2009 in the United States. HSLS:09 was made up with muliple follow-ups through 2012 to 2016. It focused on how students plan and make decisions about their education during and after high school.
